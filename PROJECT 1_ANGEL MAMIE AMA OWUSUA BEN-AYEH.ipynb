### By replacing null values with the mean of the column,the sample size can be preserved and a reasonable estimate of the missing data can be provided

### 2. DESCRIPTIVE STATISTICS
### Statistical inferences of the entire dataset

### Finding the mean for columns with null values

### Mean of fixed acidity as mean_FA

```python
mean_FA = new_dataset["fixed acidity"].mean()
mean_FA
```

### Mean of volatile aciidity as mean_VA

```python
mean_VA = new_dataset["volatile acidity"].mean()
mean_VA
```

### Mean of chlorides as mean_CL

```python
mean_CL = new_dataset["chlorides"].mean()
mean_CL
```

### Mean of free sulfur dioxide as mean_FSD

```python
mean_FSD = new_dataset["free sulfur dioxide"].mean()
mean_FSD
```

### Mean of density as mean_D

```python
mean_D = new_dataset["density"].mean()
mean_D
```

### Mean of quality as mean_Q

```python
mean_Q = new_dataset["quality"].mean()
mean_Q
```

### Inputting the mean values into the respective columns

```python
new_dataset["fixed acidity"] = new_dataset["fixed acidity"].fillna(mean_FA)
new_dataset["volatile acidity"] = new_dataset["volatile acidity"].fillna(mean_VA)
new_dataset["chlorides"] = new_dataset["chlorides"].fillna(mean_CL)
new_dataset["free sulfur dioxide"] = new_dataset["free sulfur dioxide"].fillna(mean_FSD)
new_dataset["density"] = new_dataset["density"].fillna(mean_D)
new_dataset["quality"] = new_dataset["quality"].fillna(mean_Q)
new_dataset
```

### Checking if all null values have been replaced

```python
new_dataset.isnull().any()
```

### From the information above, all null values have been replaced with the mean for their respective columns

### Line graphs indicating the quality of column in dataset
